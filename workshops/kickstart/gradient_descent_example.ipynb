## Gradient Descent Example

https://jed-ai.github.io/py1_gd_animation/

j = loss / cost function (e.g. mean squared error). In this example, we are using a quadratic function since the gradient is easier to find and plot.

w = weights / parameters / theta / beta

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_j(w):
    j = w**2 - 4*w + 2

    return j
```

```python
def gradient_descent(previous_w, learning_rate, epoch):
    w_gd = []
    j_gd = []
    
    w_gd.append(previous_w)
    j_gd.append(func_j(previous_w))

    # begin the loops to update w and j
    for i in range(epoch):
        # current_beta = prev_beta - alpha * (gradient_of_cost / beta)
        current_w = previous_w - learning_rate*(2*previous_w - 4)
        w_gd.append(current_w)
        j_gd.append(func_j(current_w))

        # update previous_w
        previous_w = current_w

    return w_gd, j_gd
```

```python
# Initialize x0 and learning rate
w0 = -0.7
learning_rate = 0.15
epoch = 10
```

```python
# j = w^2 - 4w + 2
w = np.arange(-1, 5, 0.01)
j = func_j(w)

fig, ax = plt.subplots(1, 1, sharey = True)

ax.plot(w, j, lw = 0.9, color = 'k')
ax.set_xlim([min(w), max(w)])
ax.set_ylim([-3, max(j)+1])
ax.set_xlabel('w')
ax.set_ylabel('cost')
plt.show()
```

```python
w_gd, j_gd = gradient_descent(w0, learning_rate, epoch)
```

```python
fig, ax = plt.subplots(1, 1, sharey = True)

ax.plot(w, j, lw = 0.9, color = 'k')
ax.set_xlim([min(w), max(w)])
ax.set_ylim([-3, max(j)+1])
ax.set_xlabel('w')
ax.set_ylabel('cost')

ax.scatter(w_gd, j_gd, c = 'b')

for i in range(1, epoch+1):
    ax.annotate('', xy=(w_gd[i], j_gd[i]), xytext=(w_gd[i-1], j_gd[i-1]),
    arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
        va='center', ha='center')

plt.show()
```
